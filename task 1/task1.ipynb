### Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the population data
file_path = '/content/drive/MyDrive/Prodigy internship /API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv'  # Adjust path if needed
df = pd.read_csv(file_path, skiprows=4)  # Skip metadata rows

# Step 3: Inspect the data
df.head()

# Step 4: Filter for a specific year (e.g., 2022) and drop NaNs
pop_2022 = df[['Country Name', '2022']].dropna()

# Step 5: Sort by population and take top 10 countries for visualization
top10 = pop_2022.sort_values('2022', ascending=False).head(10)

# Step 6: Create a bar chart
plt.figure(figsize=(10,6))
plt.bar(top10['Country Name'], top10['2022'], color='skyblue')
plt.title('Top 10 Most Populous Countries (2022)', fontsize=16, fontweight='bold')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/content/drive/MyDrive/Prodigy internship /API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"
df = pd.read_csv(file_path, skiprows=4)

# Filter data for the year 2022 and drop missing values
pop_2022 = df[['Country Name', '2022']].dropna()

# Sort and select top 15 countries by population
top15 = pop_2022.sort_values('2022', ascending=False).head(15)

# Plot the bar chart
plt.figure(figsize=(12, 7))
plt.bar(top15['Country Name'], top15['2022'], color='mediumseagreen')
plt.title('Top 15 Most Populous Countries (2022)', fontsize=16, fontweight='bold')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
